```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Saujanya Gadge\OneDrive\Desktop\internship\INTERNSHIP PROJECTS\Cleaned_FRvideos.csv")
data.head()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='category_name', data=data, order = data['category_name'].value_counts().index)
plt.title('Most Trending Video Categories in FR')
plt.xlabel('Number of Videos')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data['views'], bins=50, kde=True)
plt.title('Distribution of Video Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['title_sentiment'], bins=50, kde=True, color='purple')
plt.title('Sentiment Polarity Distribution of Video Titles')
plt.xlabel('Sentiment Score')
plt.ylabel('Number of Videos')
plt.show()
```

```python
plt.figure(figsize=(12,6))
data['trending_date'] = pd.to_datetime(data['trending_date'])
data.groupby('trending_date').size().plot()
plt.title('Number of Trending Videos Over Time')
plt.xlabel('Trending Date')
plt.ylabel('Number of Videos')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='views', y='likes', data=data)
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
avg_views = data.groupby('category_name')['views'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_views.values, y=avg_views.index)
plt.title('Average Views per Video Category')
plt.xlabel('Average Views')
plt.ylabel('Category')
plt.show()
```
